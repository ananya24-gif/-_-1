```python
from fastapi import FastAPI, UploadFile, HTTPException
from chromadb.client import PersistentClient
from sentence_transformers import SentenceTransformer
import asyncio

app = FastAPI()
db_client = PersistentClient()
collection = db_client.get_or_create_collection("docs_collection")
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
@app.post("/ingest/")
async def ingest_document(file: UploadFile):
    # Ensure supported file format
    if file.content_type not in ["application/pdf", "application/msword", "text/plain"]:
        raise HTTPException(status_code=415, detail="Unsupported file format")

    # Read file content (processing logic for each file type is needed)
    content = await file.read()
    # Here, parse content from DOC, DOCX, or PDF, then extract text (e.g., using PyMuPDF or python-docx)

    # Generate embeddings
    embeddings = model.encode([content])

    # Insert into ChromaDB
    doc_id = file.filename
    collection.add(embeddings=embeddings, documents=[content], metadatas=[{"id": doc_id}])
    return {"message": "Document ingested", "document_id": doc_id}
```

```python
@app.get("/query/")
async def query_document(query: str):
    # Generate embedding for the query
    query_embedding = model.encode([query])

    # Retrieve similar documents from ChromaDB
    results = collection.query(query_embedding, top_k=5)  # top_k determines the number of results
    return {"results": results}
@app.get("/query/")
async def query_document(query: str):
    # Generate embedding for the query
    query_embedding = model.encode([query])

    # Retrieve similar documents from ChromaDB
    results = collection.query(query_embedding, top_k=5)  # top_k determines the number of results
    return {"results": results}
```
